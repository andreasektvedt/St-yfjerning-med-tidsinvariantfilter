```python
import pandas as pd
import matplotlib.pyplot as plt
import os

mappe = r"C:\Users\andre\OneDrive - NTNU\1ar\ESDA 1\Designprosjekt 3"
bode_fil = f"{mappe}\\BodeBåndstoppfilter.csv"
nyttsignal_fil = f"{mappe}\\nyttlydsignalovergammeltsignal.csv"
st_fil = f"{mappe}\\s(t)_målt_på_scope.csv"
xt_fil = f"{mappe}\\x(t)inn_på_scope.csv"
spektrum_fil = f"{mappe}\\spektrumanalyseavlydsignal.csv"

# Bodeplott
df_bode = pd.read_csv(bode_fil, comment="#")
plt.figure()
plt.plot(df_bode.iloc[:, 0], df_bode.iloc[:, 1])
plt.xlabel("Frekvens (Hz)")
plt.ylabel("Forsterkning (dB)")
plt.title("Bode Båndstoppfilter - Amplitude")
plt.axvline(1017.3,color = "red")
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(mappe, "bode_magnitude.png"))



# s(t)
df_s = pd.read_csv(st_fil, comment="#")
plt.figure()
plt.plot(df_s.iloc[:, 0], df_s.iloc[:, 1])
plt.xlabel("Tid (s)")
plt.ylabel("Spenning (V)")
plt.title("s(t) målt på scope")
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(mappe, "s_t_scope.png"))

# x(t)
df_x = pd.read_csv(xt_fil, comment="#")
plt.figure()
plt.plot(df_x.iloc[:, 0], df_x.iloc[:, 1])
plt.xlabel("Tid (s)")
plt.ylabel("Spenning (V)")
plt.title("x(t) inn på scope")
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(mappe, "x_t_scope.png"))

# Nytt lydsignal over gammelt signal
df_nytt = pd.read_csv(nyttsignal_fil, comment="#")
plt.figure()
plt.plot(df_nytt.iloc[:, 0], df_nytt.iloc[:, 1], label="Spor 3")
plt.plot(df_nytt.iloc[:, 0], df_nytt.iloc[:, 3], label="Spor 5")
plt.xlabel("Frekvens (Hz)")
plt.ylabel("Forsterkning (dB)")
plt.title("Nytt lydsignal over gammelt signal - Amplitude")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(mappe, "nyttsignal_magnitude.png"))



# Spektrumanalyse av gammelt signal
df_gammelt = pd.read_csv(spektrum_fil, comment="#", encoding="latin1")
plt.figure()
plt.plot(df_gammelt.iloc[:, 0], df_gammelt.iloc[:, 1])
plt.xlabel("Frekvens (Hz)")
plt.ylabel("Forsterkning (dB)")
plt.title("Spektrumanalyse av gammelt lydsignal - Amplitude")
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(mappe, "gammelsignal_magnitude.png"))



plt.show()
```
